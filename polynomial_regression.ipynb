# Polynomil Regression
### For polynomial regression we have to use quadratic equation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# x is independent feature(randomly 100 numbers 1d form me)
x=6*np.random.rand(100,1)-3
```

```python
x
```

```python
# y is dependent feature with quadratic equation
y=0.5*x**2+1.5*x+2+np.random.rand(100,1)
```

```python
y
```

```python
print(x.shape)
print(x.shape)
```

```python
plt.scatter(x,y,color='g')
plt.xlabel('x_dataset')
plt.ylabel('y_dataset')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train,y_train,
```

```python
# implement linear regression
from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
regression_1
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
regression_1.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,regression_1.predict(x_test))
print(score)
```

```python
plt.plot(x_train,regression_1.predict(x_train),color='r')
plt.scatter(x_train,y_train)
```

```python
# polynomial
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
x_train_poly
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(x_train_poly,y_train)
score=r2_score(y_test,regression.predict(x_test_poly))
print(score)
```

```python
plt.scatter(x_train,regression.predict(x_train_poly))
plt.scatter(x_train,y_train)
```

```python
plt.plot(x_train,regression.predict(x_train_poly))
plt.scatter(x_train,y_train)
```
